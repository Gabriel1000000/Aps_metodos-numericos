```python
from visual_automata.fa.dfa import VisualDFA
```

```python

dfa = VisualDFA(
    states={"q0", "q1", "q2", "q3", "q4"},
    input_symbols={"0", "1"},
    transitions={
        "q0": {"0": "q4", "1": "q3"},
        "q1": {"0": "q2", "1": "q1"},
        "q2": {"0": "q3", "1": "q1"},
        "q3": {"0": "q0", "1": "q1"},
        "q4": {"0": "q0", "1": "q2"},
    },
    initial_state="q0",
    final_states={"q1"},
)

```

```python
dfa.show_diagram()
```

```python
#Exemplo de conversão de base
numero_base10 = 443
numero_base2 = bin(numero_base10)
numero_base8 = oct(numero_base10)
numero_base16 = hex(numero_base10)
print(f'numero_base10 = {numero_base10}')
print(f'numero_base2 = {numero_base2}')
print(f'numero_base8 = {numero_base8}')
print(f'numero_base16 = {numero_base16}')

print(f'{numero_base2} na base 10 = {int(numero_base2,2)}')
print(f'{numero_base8} na base 10 = {int(numero_base8,8)}')
print(f'{numero_base16} na base 10 = {int(numero_base16,16)}')
```
